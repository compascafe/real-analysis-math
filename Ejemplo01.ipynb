```python
import math
import matplotlib.pyplot as plt
```

```python

def compute_sine_taylor(x, num_terms):
    approximation = 0.0
    
    for n in range(num_terms):
        numerator = (-1) ** n * x ** (2 * n + 1)
        denominator = math.factorial(2 * n + 1)
        approximation += numerator / denominator
        
    return approximation

def plot_sine_taylor_approximation():
    x_values = [x / 10 for x in range(-50, 51)]  # Generate x values from -5 to 5
    
    # Compute sine values using math.sin for comparison
    sine_values = [math.sin(x) for x in x_values]
    
    # Compute Taylor series approximations
    taylor_values = [compute_sine_taylor(x, 5) for x in x_values]
    
    # Plot the results
    plt.plot(x_values, sine_values, label='sin(x)')
    plt.plot(x_values, taylor_values, label='Taylor (5 terms)')
    plt.xlabel('x')
    plt.ylabel('Value')
    plt.title('Taylor Series Approximation of sin(x)')
    plt.legend()
    plt.show()

# Call the function to generate the plot
plot_sine_taylor_approximation()
```
